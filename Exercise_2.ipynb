```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load the Iris dataset as a DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Initialize a list to store sample means
sample_means = []

# Perform random sampling 100 times
for _ in range(100):
    # Randomly sample 30 observations
    sampled_df = df.sample(n=30)
    # Calculate the mean of the 'sepal length (cm)' column for the sample
    sample_mean = sampled_df['sepal length (cm)'].mean()
    # Append the mean to the list
    sample_means.append(sample_mean)

# Display the sample means
print(sample_means)

# Calculate overall statistics of the sample means
print(f"Mean of sample means: {np.mean(sample_means)}")
print(f"Standard deviation of sample means: {np.std(sample_means)}")
```

evaluates how much sample means vary due to random sampling.
It helps in understanding the behavior of sample statistics.

```python
import matplotlib.pyplot as plt
plt.hist(sample_means, bins=15, edgecolor='black', alpha=0.7)
plt.title('Distribution of 100 Sample Means (Sepal Length)')
plt.xlabel('Sample Mean (Sepal Length in cm)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define a function for systematic sampling
def systematic_sampling(dataframe, step):
    indices = np.arange(0, len(dataframe), step)
    sampled_df = dataframe.iloc[indices]
    return sampled_df

# Perform systematic sampling (20% of the dataset)
step_size = int(len(df) / (len(df) * 0.2))  # Calculate step size for 20% sampling
systematic_sample = systematic_sampling(df, step_size)

# Plot original dataset and sampled dataset
plt.figure(figsize=(12, 6))

# Original dataset
plt.subplot(1, 2, 1)
plt.scatter(df.index, df['sepal length (cm)'], color='blue', alpha=0.6, label='Original Data')
plt.title('Original Dataset')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()

# Sampled dataset
plt.subplot(1, 2, 2)
plt.scatter(systematic_sample.index, systematic_sample['sepal length (cm)'], color='red', alpha=0.8, label='Systematic Sample')
plt.title('Systematic Sample (20%)')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()

plt.tight_layout()
plt.show()
```
