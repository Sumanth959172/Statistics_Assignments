Reading Dataset From Kaggle

```python
import pandas as pd
import os

# Assuming the CSV file is named 'tested.csv' and is inside the '6' directory
file_path = os.path.join("/root/.cache/kagglehub/datasets/brendan45774/test-file/versions/6", "tested.csv")

# Now read the CSV using the corrected file path
df = pd.read_csv(file_path)
print(df.head())
```

Measure of Central Tendency

```python
for column in df.select_dtypes(include=['number']).columns:
    mean = df[column].mean()
    median = df[column].median()
    mode = df[column].mode()[0]  # Mode might return multiple values, taking the first one

    print(f"Central Tendency for column '{column}':")
    print(f"  Mean: {mean}")
    print(f"  Median: {median}")
    print(f"  Mode: {mode}")
```

```python
import numpy as np
for column in df.select_dtypes(include=['number']).columns:
    data_range = np.ptp(df[column])  # Calculate range using numpy.ptp
    std_dev = df[column].std()
    variance = df[column].var()

    print(f"Measure of Dispersion for column '{column}':")
    print(f"  Range: {data_range}")
    print(f"  Standard Deviation: {std_dev}")
    print(f"  Variance: {variance}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
for column in df.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 6))  # Adjust figure size as needed
    sns.histplot(df[column], kde=True)  # kde adds a kernel density estimate
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```
